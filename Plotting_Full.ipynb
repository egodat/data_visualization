## Plotting coefficients(A) plots

These plots show how the coefficients derived from fit parameters evolve as a function of A, the atomic mass of the nuclei. There are two functions here, one to govern plotting and one to describe the A-dependence of the coefficients.

### Functions

```python
def param_A_dep(paramlist, c ,A):
    
    l = len(paramlist)/3
    paramsplit=[paramlist[0*l:1*l],paramlist[1*l:2*l],paramlist[2*l:3*l]]
    
    paramkeys=["p","a","b"]
    paramval=[1.0,1.0,1.0]
    
    #return paramsplit
    for n in range(3):
        paramkey = paramkeys[n]
        for i in range(len(paramsplit[n])):
            if paramsplit[n][i][0][-1]==str(c):
                if paramsplit[n][i][0][-2]==paramkey:
                    paramval[n]=paramsplit[n][i][1]
                    #print paramval[n]
                else: print "Unknown Parameter name"

    #value = (paramval[0]+paramval[1]*(1.0-A**(-paramval[2])))/paramval[0]
    value = 1.0+(paramval[1]*(1.0-A**(-paramval[2])))/paramval[0]
    
    return value
```

```python
def plotfunc_cA(param,flavlist,savefigure=False,name='test'):
    
    plt.figure()
    colorlist=["k","b","r","g","m","c"]
    stylelist=["solid","dashed","dotted","solid","dashed","dotted"]
    paramname={'gps':'{g}','uvps':'{u_v}','dvps':'{d_v}','ubdbps':r'{\bar{u}+\bar{d}}','ssbps':r'{s+\bar{s}}'}
    xvals=np.array(range(1,200))
    for val in range(len(flavlist)/3):
        plt.plot(xvals,param_A_dep(flavlist, val, xvals),linewidth=1,color=colorlist[val],linestyle=stylelist[val],label="$c^"+paramname[param]+"_"+str(val)+"$")
    
    #plt.legend(loc='upper left')
    plt.legend(bbox_to_anchor=(0., 1.005, 1., .102), loc=3,ncol=6, mode="expand", borderaxespad=0.)
    #plt.legend(bbox_to_anchor=(1.02, 1), loc=2, ncol=1,borderaxespad=0.)
    ylabel(r"$\frac{p_k+a_k(1-A^{-b_k})}{p_k}$",fontsize=19)
    xlabel(r"$A$",fontsize=19)
    title(r"$"+paramname[param]+"$",fontsize=19)
    plt.tight_layout()
    
    if savefigure:
        print "EXPORTING: "+"results/fits/coeffplot/" + name + param + ".pdf"
        savefig("results/fits/coeffplot/" + name + param + ".pdf")
```

## Plotting  $\chi^2$ / dof per exp

These plots describe the $\chi ^2$ per degree of freedom for reach experiment included in the fit.

### Functions

```python
def plot_chi2_per_data(name,comboarray,save=False, rmZero=True):
    
    if rmZero:
        comboarray=np.transpose(comboarray)
        tmp=[]
        for i in range(len(comboarray)):
            if comboarray[i][1]==0:
                #print "Found 0"
                tmp=np.append(tmp,i)
        comboarray=np.transpose(np.delete(comboarray,tmp,0))
    
    
    chisq = comboarray[2]   

    chisq_series = pd.Series.from_array(chisq)   

    x_labels = comboarray[0]
    x_labels=[int(el) for el in x_labels]

    colorlist=[]
    for i in range(len(comboarray[0])):
        if comboarray[0][i]<5200 and comboarray[0][i]>=5100:
            colorlist=np.append(colorlist,"b")
        if comboarray[0][i]<6200 and comboarray[0][i]>=5200:
            colorlist=np.append(colorlist,"r")
        if comboarray[0][i]>6200:
            colorlist=np.append(colorlist,"g")

    plt.figure(figsize=(12, 8))
    ax = chisq_series.plot(kind='bar',color=colorlist)
    ax.set_title("")
    ax.set_xlabel("$Set~ID$",fontsize=19)
    ax.set_ylabel("$\chi^2/d.o.f.$",fontsize=19)
    ax.set_xticklabels(x_labels)
    plt.grid(axis='both')
    rects = ax.patches
    
    plt.ylim(0,6)

    info_patch = [mpatches.Patch(color="b",label="DIS"),mpatches.Patch(color="r",label="DY"),mpatches.Patch(color="g",label="W/Z")]
    ax.legend(handles=info_patch,loc='upper left')

    labels = comboarray[1]
    labels=[int(el) for el in labels]

    for rect, label in zip(rects, labels):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2, height, label, ha='center', va='bottom')
        
    if save:
        print 'Exporting: '+"results/fits/chi2perdata/"+"chi2perdof_" + name + ".pdf"
        savefig("results/fits/chi2perdata/"+"chi2perdof_" + name + ".pdf")
```

## Plot multiple PDF for all A

Plot PDFs for several values of A on a single bandplot. This requires the LHAPDF python interface to be installed.

### Function

```python
def plot_bandplot(pdfarray,flavor="g",qval=10,Xrange=(10**-3,1.0),save=False,stem="test"):
    
    flavsave=flavor
    
    xmin=np.log10(Xrange[0])
    xmax=np.log10(Xrange[1])

    xlist=np.logspace(xmin, xmax, num=150)

    flav={'ub':-2,"db":-1,"g":0,"d":1,"u":2,"s":3}

    flavname={"ub":r'{\bar{u}}',"db":r'{\bar{d}}',"g":r'{g}',"d":r'{d}',"u":r'{u}',"s":r'{s}',"uv":r'{u_v}',"dv":r'{d_v}'}
    for i in range(len(pdfarray)):

        pdfatx = []
        if flavsave=="uv":
            for x in xlist:
                pdfatx = append(pdfatx,pdfarray[i].xfxQ2(flav["u"], x, qval**2)-pdfarray[i].xfxQ2(flav["ub"], x, qval))
        if flavsave=="dv":
            for x in xlist:
                pdfatx = append(pdfatx,pdfarray[i].xfxQ2(flav["d"], x, qval**2)-pdfarray[i].xfxQ2(flav["db"], x, qval))
        elif flavsave in flav:
            for x in xlist:
                pdfatx = append(pdfatx,pdfarray[i].xfxQ2(flav[flavsave], x, qval**2))


        ######### PLOTTING ###############
        if np.mod(i,3)==0:
            sty="solid"
        elif np.mod(i,3)==1:
            sty="dashed"
        else:
            sty="dotted"
    
        plt.plot(xlist,pdfatx,label="$A="+nucarray[i].split("_")[1]+"$",linestyle=sty)
        plt.xscale('log')
                
        lgd=plt.legend(bbox_to_anchor=(1.002, 1.00), loc=2, ncol=1,borderaxespad=0.,fontsize=10)
        ylabel(r"$xf^{p/A}$",fontsize=19)
        xlabel(r"$x$",fontsize=19)
        title("$"+flavname[flavsave]+"$",fontsize=19)
        plt.tight_layout()
        
    if save:
        if qval==1.3:
            qname="q0"
        else:
            qname=str(qval)
        print "Exporting: "+"results/fits/bandplots/" + stem + "_A_" + flavsave + "_" +qname+".pdf"
        savefig("results/fits/bandplots/" + stem + "_A_" + flavsave + "_" +qname+".pdf", bbox_extra_artists=(lgd,), bbox_inches='tight')
```

## Data vs Theory

These plots put the data with error bars and the theory predictions for each data set on individual plots. There are a lot of different functions here all designed to handle different data sets and different theory types. All of these different functions are wrapped by a master switching function "plot_data_set". Should any new functions be added in the future, this will need to be modified.

### Functions

```python
def nuc_lookup(a,z):
    periodicT={'p': (1,1),'n': (1,0),'D': (2,1),'He_{i}': (3,1.5),'He': (4,2),'Li_{i}': (6,3),'Li': (7,3),'Be_{i}': (9,4.5),'Be': (9,4), 'C': (12,6), 'N': (14,7), 'Al': (27,13),'Al_{i}': (27,13.5), 'Ca': (40,20), 'Fe': (56,26), 'Fe_{i}': (56,28), 'Cu': (64,32), 'Kr_{i}': (84,42), 'Ag': (108,54), 'Sn_{i}': (119,59.5), 'Xe': (131, 54), 'W': (184,74), 'Au': (197,98.5), 'Pb': (208,82), 'Pb_{87}': (208,87), 'Pb_{i}': (207,103.5)}
    for key in periodicT: 
        if periodicT[key][0] == a:
            if periodicT[key][1]==z:
                return key
    return "Unknown"
```

```python
def plot_data_set(inputFile,setID,save=False):
    
    # Crop list of Points to just specific Data Set
    datapts=[]
    for el in inputFile["Chi2AndDataTable"]:
        if el["IDDataSet"]==setID:
            datapts=np.append(datapts,el)
    if len(datapts)==0:
        print "No points found for that Set ID"
        return datapts
    
    #Determine the type of point
    theoType=datapts[0]["TypeTheo"]
    expType=datapts[0]["TypeExp"]
    
    if theoType == "F2":
        return plot_theo_f2(datapts,save)
    if expType=="DIS" and theoType=="RSIG" or setID == 5159:
        return plot_theo_rsigDIS(datapts,save)
    if theoType == 'RF2':
        return plot_theo_rf2(datapts,save)
    if theoType == "RSIG":
        return plot_theo_rsig(datapts,save)
    if theoType == "APPLGRID":
        return plot_theo_wzprod(datapts,save)
    else:
        print"Unknown Theory Type:"
        return theoType
```

```python
def plot_theo_rf2(pts,save=False,plotX=False):
    a1,z1 = pts[0]['A1'], pts[0]['Z1']
    a2,z2 = pts[0]['A2'], pts[0]['Z2']

    nuc1 = nuc_lookup(a1,z1)
    nuc2 = nuc_lookup(a2,z2)
        
    typeExp = pts[0]['TypeExp']
    setnum = pts[0]['IDDataSet']
    
    points=[]
    for el in pts:
        tmp = (el['Chi2'],el["Theo"],el["Data"],el["Stats"],el["X"],el["Q2"]) #creates a list of information for plotting (chi2,theory,data,error,x,q2)
        points.append(tmp)
    #print np.transpose(points)[5],np.transpose(points)[3]
    
    chi2 = np.round(sum(np.transpose(points)[0]),3)
    
    fig = plt.figure(0)
    #plt.plot(np.transpose(points)[4],np.transpose(points)[1])
    plt.xscale('log')
    
    if plotX:
        plt.xlim(10**-3, 1)
        #xlabel(r"$x$",fontsize=19)
        xstring = "x"
        x = np.transpose(points)[4]
    else:
        plt.xlim(1, 1000)
        #xlabel(r"$Q^2$",fontsize=19)
        xstring = "Q^2"
        x = np.transpose(points)[5]
    #Horizontal Line at y=1
    plt.axhline(1.0, color='k', linestyle='--')
    #Theory Points Joined with line    
    y = np.transpose(points)[1]
    theory=plt.plot(x, y,marker="o",linewidth=2,color="r",label="Theory")
    #Data Points with Errors
    y = np.transpose(points)[2]
    err = np.transpose(points)[3]
    data=plt.plot(x, y, marker="s",linewidth=0,color="b",label="Data")
    plt.errorbar(x, y, yerr=np.sqrt(err),fmt='none',elinewidth=2,color="b")
    
    lgd2=plt.legend(loc="upper left")
    plt.gca().add_artist(lgd2)
    
    plt.ylim(0.75, 1.25)
    
    patchlabel=r"SetID: "+str(setnum)+"   "+"ExpType: "+typeExp+"   "+"Obs: $F_2^{"+nuc1+"}/F_2^{"+nuc2+"}("+xstring+")$"+"   "+"$\chi^2: "+str(chi2)+"$"+"   "+"npts: $"+str(len(x))+"$"
    
    #info_patch = mpatches.Patch(color="w",label=patchlabel)
    info_patch = mpatches.Patch(visible=False,label=patchlabel)
    lgd=plt.legend(handles=[info_patch],bbox_to_anchor=(0., 1.005, 1., .102), loc=3, mode="expand", borderaxespad=0.)
    
    


    plt.tight_layout()
    
    if save:
        savefig("results/fits/dataplots/datavstheory_" + typeExp + "_" + str(setnum) + ".pdf", bbox_extra_artists=(lgd,lgd2,), bbox_inches='tight')
```
